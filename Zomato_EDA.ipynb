```python
# Now we will import the other file
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
#assign variable to create pie chart
country_names=final_df.Country.value_counts().index 
```

```python
country_val=final_df.Country.value_counts().values
```

## pie chart to see which country has maximum records

```python
fig=go.Figure(go.Pie(labels=country_names[:3],values=country_val[:3]))
fig.show()
```

## observations :
Zomato has maximum records are from India After that United States and then uk

## Zomato Ratings 

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head() 
```

## observations
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Rating is between 4.0 to 4.4 ---> Very good
3. When Rating is between 3.5 to 3.9 ---> Good
4. When Rating is between 2.5 to 3.4 ---> Average
5. When Rating is between 1.8 to 2.4 ---> Poor
6. There are 2148 people that heven't Rated.

## Bar chart on Ratings

```python
ratings.iplot(kind = 'bar', x = 'Aggregate rating', y = 'Rating Count', title = 'ratings')
```

## observations: 
    
    1. Not Rated Count is Very High.
    2. Maximum Number of rating are between 2.5 to 3.4.
